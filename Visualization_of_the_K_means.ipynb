#Visualization of the K-means


```python
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

## Генерация групп точек

```python
k = 3
n = 1000

x = [[[random.normalvariate(-2+(j+1)*1,0.7), random.normalvariate(-3+(j+0.4)*3*(-1)**j,0.8)] \
      for i in range(n)] for j in range(k)]
X = np.array(x[0]+x[1]+x[2])
```

### Вывод рандомных групп точек на график

```python
fig, ax = plt.subplots(figsize=(12, 10))
msize = 4.5

plt.plot (X[:, 0], X[:, 1], marker='o', linestyle='', markersize=msize)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Запуск k-means при k=2

```python
k_klasses = 2
kmeans = KMeans(k_klasses)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```python
plt.figure(figsize = (12,10))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=25, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=150, alpha=0.5, label= 'центры классов');

plt.legend()
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
